# Computer OS Market Share

Everyone who uses a computer uses an operating system and most of us are obsessed with the OS we use (just like sports fans are obsessed over their teams). Everyone tries to highlight how their OS is better than the rest. It is almost like a war out there.

And observing the market share is a good way to see how each operating system is performing in the OS wars.

For that purpose we have this tiny dataset:
* 10 rows, 4 columns in the dataset
* the data is observed from January 2014 to December 2022

Although having a bigger dataset would be better, but we can get things done with the dataset at hand also.

# Importing and Observing the Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('compOsShare.csv')
print(df.info())
print("-" * 50)
print(df.head())
```

1. Month should be `int`.
2. Rest of the columns should be `float`.
3. There should be no `%` signs at the end of observations for each OS.
4. The order of dataset should be in reverse (from `2014` to `2022` not from `2022` to `2014`).
5. The column names end with a blankspace, we need to remove that.

# Doing some Data Validation

```python
df.columns = ['Month','Windows', 'MacOs', 'Linux']
print(df)
```

```python
df['Windows'] = df['Windows'].str.replace('%', '')
df['MacOs'] = df['MacOs'].str.replace('%', '')
df = df.iloc[::-1, :]
print(df)
```

```python
for col in ['Windows', 'MacOs', 'Linux']:
    df[col] = df[col].astype(float)

print(df.info())
```

# The Plot (Graph)

```python
sns.set_theme(style='darkgrid', palette='Dark2')
for col in ['MacOs', 'Linux']:
    sns.lineplot(x='Month', y=col, data=df, legend='auto')
    
#plt.axes()
plt.ylabel('MacOs, Linux Market Share %')
plt.legend(['Mac_Os', 'Linux'])
plt.xticks(rotation=90)
plt.show()

sns.lineplot(x='Month', y='Windows', data=df)
plt.ylabel('Windows Market Share %')
plt.xticks(rotation = 90)
plt.show()
```

# Windows Observation

* Windows is dominating the market, it is like people are only using windows.
* There is a incline in Windows Market share till 2017.
* A sharp drop in year 2018.
* A steady decline from December 2019 (from the start of COVID and the Windows 11 update).

# Mac and Linux (Unix based) Observations

* Linux rising steadliy.
* MacOs a slight decline.
* See a sharp decline in 2018.
* In 2022 MacOs at 2.5% market share.
* In 2022 Linux at almost 1.4% market share.
